 # find and find_all

```python
style = soup.find('style') #finds the first occurence in the content
style
```

```python
#finds all the style tags in our content
all_styles = soup.find_all('style')
all_styles
```

PASS IN A LIST OF ELEMENTS TO LOOK FOR

```python
# If you pass a list of elements it returns the first element of the list it founds
l = soup.find(['h4','h2'])
l
```

```python
# finds all the events of all elements in our list
l = soup.find_all(['h4','h2'])
l
```

# 𝐐) 𝐒𝐜𝐫𝐚𝐩𝐞 𝐭𝐡𝐞 𝐭𝐚𝐛𝐥𝐞 𝐟𝐫𝐨𝐦 𝐭𝐡𝐞 𝐰𝐞𝐛𝐩𝐚𝐠𝐞?

```python
table = webpage.select('.hockey-stats')[0]
table
```

```python
import pandas as pd

columns = table.find_all('th')
columns
```

```python
column_name = []
for i in columns:
    column_name.append(i.string)
```

```python
column_name
```

```python
df = pd.DataFrame(columns=column_name)
```

```python
df
```

```python
table.find_all('td')
```

```python
table_rows = table.find_all('tr',attrs={'class':'team-continent-NA'})
table_rows
```

```python
l = []
for row in table_rows:
    td = row.find_all('td')
    row_data = [(tr.get_text()).strip() for tr in td]
    l.append(row_data)
```

```python
l
```

```python
for i in range(len(l)):
    df.loc[i] = l[i]
```

```python
df
```
