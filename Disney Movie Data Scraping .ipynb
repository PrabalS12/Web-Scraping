```python
def loading_content(url):
        r = requests.get('https://en.wikipedia.org'+url)
        film = bs(r.content)
        return film
   
def content_val(t_row_data):
    if t_row_data.find('li'):
        return [i.get_text(" ",strip=True).replace('\xa0'," ") for i in t_row_data.find_all('li')]
    # FOR SEPARATING THE LONG STRINGS WHERE NAMES WEREN'T SEPARATED
    # at every break tag (a/c to inspect) we strip that text
    # and make it an element in list
    elif t_row_data.find('br'):
        return [text for text in t_row_data.stripped_strings]
    else:
        return t_row_data.get_text().replace('\xa0'," ")

# FUNCTION TO REMOVE THE "SUP" & "SPAN" TAG FROM OUR SELECTION
# to remove references i.e sup tag, ex: [1], [2], etc
# to remove extra info in release date i.e span tag, ex: (19.05.2022)
def clean_tags(film):
    for tag in film.find_all(['sup', 'span']):
        tag.decompose()
    
        
def scraping_webpage(url):
    r = requests.get(url)
    film = bs(r.content)
    table = film.find(class_='infobox vevent')
    trows = table.find_all('tr')
    
    clean_tags(film)
    movie_info = {}
    for index,row in enumerate(trows):
        if index==0:
            movie_info['Title'] = row.find('th').get_text(" ",strip=True)
        else:
            header = row.find('th')
            if header:
                key = row.find('th').get_text(" ",strip=True)
                val = content_val(row.find('td'))
                movie_info[key] = val 
    return movie_info    
```

```python
def running_time_to_int(movies_list):
    for i in movies_list:
        try:
            if type(i['Running time']) == list:
                i['Running time'] = int((max(i['Running time'])).split(" ")[0])
            else:
                i['Running time'] = int(i['Running time'].split(" ")[0])
        except Exception as e:
            print(e)
            i['Running time'] = 0
```

```python
def budget_to_int(i):   
        try:
            s = i['Budget'] 
            if type(s) == str:
                
                if '-' in s:
                    i['Budget'] = s.split(" ")[0].split('$')[-1].strip().split('-')[-1]
                
                elif "—" in s:
                    i['Budget'] = s.split(" ")[0].split('$')[-1].strip().split('—')[-1]
                
                elif "–" in s:
                    i['Budget'] = s.split(" ")[0].split('$')[-1].strip().split('–')[0]
                    
                elif ('million' or 'Million') not in s:
                    i['Budget'] = int("".join(s.split('$')[1].split(',')))/1000000
                
                else:
                    i['Budget'] = s.split("million" or "Million")[0].split('$')[1].strip().split('to')[-1]
            
            
            elif type(s) == list:
                for j in s: 
                    if ("$") in j:
                        i['Budget'] = j.split("million" or "Million")[0].split('$')[1]
                    elif "crore" in j:
                        i['Budget'] = j.split('crore')[0]
        
        except Exception as e:
            i['Budget'] = 'NA'
            print(e)
```

```python
def boxoffice_to_int(i):
        try:
            s = i['Box office']
            if type(s) == str:
                if '-' in s:
                    i['Box office'] = s.split(" ")[0].split('$')[-1].strip()
                    
                elif "–" in s:
                    i['Box office'] = s.split(" ")[0].split('$')[-1].strip().split("–")[-1]

                elif 'crore' in s:
                    i['Box office'] = s.split("crore")[0]
               
                elif 'billion' in s:
                    i['Box office'] = float(s.split("billion")[0].split('$')[1])*1000
                
                elif ('million' or 'Million') not in s:
                    i['Box office'] = int("".join(s.split('$')[1].split(',')).split(" ")[0])/1000000
                
                else:
                    i['Box office'] = s.split("million" or "Million")[0].split('$')[1].strip()

            elif type(s) == list:
                x = 0
                for j in s: 
                    if "$" in j and ',' not in j:
                        if x < float((j.split("million" or "Million")[0].split('$')[1])):
                            x = float(j.split("million" or "Million")[0].split('$')[1])
                i['Box office'] = x
                
        except Exception as e:
            i['Box office'] = 'NA'
            print(e)
```

```python
# Another approach using regex library

import re

amounts = r"thousand|million|billion"
number = r"\d+(,\d{3})*\.*\d*"

word_re = rf"\${number}(-|\sto\s|–)?({number})?\s({amounts})"
value_re = rf"\${number}"

def word_to_value(word):
    value_dict = {"thousand": 1000, "million": 1000000, "billion": 1000000000}
    return value_dict[word]

def parse_word_syntax(string):
    value_string = re.search(number, string).group()
    value = float(value_string.replace(",", ""))
    word = re.search(amounts, string, flags=re.I).group().lower()
    word_value = word_to_value(word)
    return value*word_value

def parse_value_syntax(string):
    value_string = re.search(number, string).group()
    value = float(value_string.replace(",", ""))
    return value

'''
money_conversion("$12.2 million") --> 12200000 ## Word syntax
money_conversion("$790,000") --> 790000        ## Value syntax
'''
def money_conversion(money):
    if money == "N/A":
        return None

    if isinstance(money, list):
        money = money[0]
        
    word_syntax = re.search(word_re, money, flags=re.I)
    value_syntax = re.search(value_re, money)

    if word_syntax:
        return parse_word_syntax(word_syntax.group())

    elif value_syntax:
        return parse_value_syntax(value_syntax.group())

    else:
        return None
```

```python
from datetime import datetime

dates = [movie.get('Release date', 'N/A') for movie in movie_info_list]

def clean_date(date):
    return date.split("(")[0].strip()

def date_conversion(date):
    if isinstance(date, list):
        date = date[0]
        
    if date == "N/A":
        return None
        
    date_str = clean_date(date)

    fmts = ["%B %d, %Y", "%d %B %Y"]
    for fmt in fmts:
        try:
            return datetime.strptime(date_str, fmt)
        except:
            pass
    return None
```

```python
for movie in movie_info_list:
    movie['Release date (datetime)'] = date_conversion(movie.get('Release date','N/A'))
```

# Making api requests to fetch data and add ratings into our dataset

```python
import requests
import urllib

def get_api_info(title):
    base_url = 'http://www.omdbapi.com/?'
    parameters = {'apikey':'30b17724','t':title}
    para_encoded = urllib.parse.urlencode(parameters)
    full_url = base_url + para_encoded
    return requests.get(full_url).json()

def get_rotten_tomato(api_info_dict)'':
    rating = api_info_dict.get('Ratings', [])
    for score in rating:
        if score['Source'] == 'Rotten Tomatoes':
            return score['Value']
    return None   
```

```python
get_api_info('Into the woods')
```

```python
for movie in movie_info_list:
    api_dict = get_api_info(movie['Title'])
    rated = get_rotten_tomato(api_dict)
    movie['IMDB ratings'] = api_dict.get('imdbRating',None)
    movie['Rotten Tomatoes Ratings'] = rated
```
